## üìà Numerical Data Plotting

Numerical data plotting is used to explore relationships between continuous variables in a dataset.  
It helps in identifying trends, patterns, clusters, and possible correlations between numerical features.

Seaborn provides high-level functions that make numerical data visualization simple and visually appealing.  
In this section, we begin with **relplot()**, which is used to uncover relationships between numerical variables.


### üîπ relplot(): Uncovering Relationships Between Numerical Variables

The `relplot()` function in Seaborn is a high-level interface used to visualize the relationship between two numerical variables.  
It is commonly used to create scatter plots or line plots and is useful for identifying how one variable changes with respect to another.

In this visualization, we analyze the relationship between:
- **carat**: Weight of the diamond  
- **price**: Price of the diamond in USD

This helps us understand how diamond price varies with carat weight.


```python
g = sns.relplot(
    x="carat",
    y="price",
    data=df,
    kind="scatter"
)

# Set title
g.fig.suptitle("Relationship Between Carat and Price", y=1.02)

# Save the figure
g.savefig("relplot_carat_price.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The plot shows a clear positive relationship between **carat** and **price**.  
As the carat weight of a diamond increases, the price also tends to increase significantly.

The scatter pattern indicates:
- Higher carat diamonds are generally more expensive
- The relationship is non-linear, with price increasing sharply for higher carat values

This visualization confirms that **carat is a strong influencing factor in determining diamond price**.


### üîπ scatterplot(): Visualizing Numerical Relationships

The `scatterplot()` function in Seaborn is used to visualize individual data points for two numerical variables.  
It helps in identifying correlations, clusters, and outliers in the data.

In this plot, we visualize the relationship between:
- **carat**: Weight of the diamond  
- **price**: Price of the diamond  

This scatter plot provides a clearer view of how diamond prices vary with carat weight.


```python
plt.figure(figsize=(8, 5))

sns.scatterplot(
    x="carat",
    y="price",
    data=df,
    hue="price",          # gradient based on price
    palette="viridis",    # smooth gradient color map
    legend=False          # hide legend to keep plot clean
)

plt.title("Scatter Plot of Carat vs Price (Gradient by Price)")

# Save the plot
plt.savefig("scatterplot_carat_price_gradient.png", dpi=300, bbox_inches="tight")

plt.show()
```

The color gradient enhances the visualization by indicating price intensity.
Darker shades represent higher-priced diamonds, making it easier to observe how price increases with carat weight.


### üîπ lineplot(): Highlighting Data Trends

The `lineplot()` function in Seaborn is used to visualize trends in data by connecting data points with a line.  
It is especially useful for identifying overall patterns and trends across a continuous variable.

In this visualization, we use a line plot to observe the general trend between:
- **carat**: Weight of the diamond  
- **price**: Price of the diamond  

Although the Diamonds dataset is not time-based, the line plot helps in understanding how price changes as carat increases.


```python
# Create carat bins for gradient effect
df["carat_bin"] = pd.qcut(df["carat"], q=5)

plt.figure(figsize=(8, 5))

sns.lineplot(
    x="carat",
    y="price",
    data=df,
    hue="carat_bin",
    palette="viridis",
    errorbar=None
)

plt.title("Line Plot Showing Price Trend Across Carat Ranges")

# Save the plot
plt.savefig("lineplot_carat_price_gradient.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The line plot with a color gradient highlights how diamond prices change across different carat ranges.  
Lighter shades represent lower carat ranges, while darker shades represent higher carat ranges.

Key insights:
- Price increases consistently across all carat bins.
- Higher carat ranges show steeper price trends.
- The gradient helps visually distinguish pricing behavior across diamond sizes.

This visualization enhances trend interpretation by combining numerical trends with categorical grouping.


Note: Line plots do not support continuous color gradients directly.
To achieve a gradient effect, numerical values are grouped into bins and visualized using a color palette.

## üìä Categorical Data Plotting

Categorical data plotting is used to compare numerical values across different categories.

In this section, we explore categorical data visualization using:

We begin with `catplot()`, which is a high-level interface for drawing categorical plots.


### üîπ catplot(): Visualizing Categorical Data

The `catplot()` function in Seaborn is a versatile, high-level function used to visualize relationships between categorical and numerical variables.  
It supports different plot types such as bar, box, count, and point plots.

In this visualization, we analyze how the **average price of diamonds varies across different cut categories**.


```python
g = sns.catplot(
    x="cut",
    y="price",
    hue="cut",
    data=df,
    kind="bar",
    palette="viridis",
    legend=False
)

g.fig.suptitle("Average Diamond Price by Cut", y=1.02)

# Save the plot
g.savefig("catplot_price_by_cut.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

### üîπ boxplot(): Revealing Data Distribution and Outliers

The `boxplot()` function in Seaborn is used to visualize the distribution of numerical data across different categories.  
It displays the median, quartiles, and potential outliers, making it useful for comparing distributions between groups.

In this visualization, we analyze the distribution of **diamond prices across different cut categories** to understand how cut quality affects price variation.


```python
plt.figure(figsize=(8, 5))

sns.boxplot(
    x="cut",
    y="price",
    data=df,
    hue="cut",
    palette="viridis",
    legend=False
)

plt.title("Distribution of Diamond Prices by Cut")

# Save the plot
plt.savefig("boxplot_price_by_cut.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The boxplot reveals the distribution and spread of diamond prices across different cut categories.

Key insights:
- **Ideal** and **Premium** cuts show higher median prices compared to other cuts.
- The presence of numerous outliers indicates wide price variation within each cut category.
- Diamonds with lower cut quality tend to have lower median prices but still exhibit some high-priced outliers.

This visualization highlights both central tendency and variability, making it effective for comparing price distributions across categories.


### üîπ stripplot(): Visualizing Individual Data Points

The `stripplot()` function in Seaborn is used to display individual data points along a categorical axis.  
It is useful for observing the distribution and density of data points across different categories.

In this visualization, we examine the distribution of **diamond prices across different cut categories** by plotting each individual data point.


```python
plt.figure(figsize=(8, 5))

sns.stripplot(
    x="cut",
    y="price",
    data=df,
    hue="cut",
    palette="viridis",
    jitter=True,
    legend=False
)

plt.title("Strip Plot of Diamond Prices by Cut")

# Save the plot
plt.savefig("stripplot_price_by_cut.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The strip plot shows individual diamond prices for each cut category.

Key observations:
- A dense concentration of data points appears at lower price ranges for all cut types.
- Higher cut categories exhibit a wider spread of prices.
- Jitter helps in reducing overlap and makes individual data points more visible.

This plot provides a detailed view of price distribution at the individual data level.


## üìä Visualizing Data Distribution

Understanding data distribution is essential for identifying patterns such as skewness, spread, and concentration of values.  
Seaborn provides multiple functions to visualize how numerical data is distributed.

In this section, we explore data distribution using:
- Histogram (distplot replacement)
- KDE plot
- Joint plot
- Rug plot

We begin with a histogram to analyze the distribution of diamond prices.


### üîπ distplot(): Analyzing Data Distribution

The `distplot()` function was traditionally used to visualize data distribution using histograms and density curves.  
Since it is deprecated, we use `histplot()` with KDE enabled, which provides the same functionality.

In this visualization, we analyze the distribution of **diamond prices** to understand how prices are spread across the dataset.


```python
plt.figure(figsize=(8, 5))

sns.histplot(
    df["price"],
    kde=True,
    color="purple"
)

plt.title("Distribution of Diamond Prices")

# Save the plot
plt.savefig("histplot_price_distribution.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The histogram shows that diamond prices are **right-skewed**.

Key observations:
- Most diamonds fall within the lower price range.
- A small number of diamonds have very high prices, forming a long tail.
- The KDE curve highlights the overall shape and skewness of the distribution.

This indicates that high-priced diamonds are relatively rare compared to lower-priced ones.


### üîπ kdeplot(): Kernel Density Estimation

The `kdeplot()` function in Seaborn is used to visualize the probability density of a continuous numerical variable.  
It provides a smooth curve that represents the distribution of data, making it easier to identify peaks, skewness, and spread.

In this visualization, we use a KDE plot to analyze the distribution of **diamond prices**.


```python
plt.figure(figsize=(8, 5))

sns.kdeplot(
    df["price"],
    fill=True,
    cmap="viridis"
)

plt.title("KDE Plot of Diamond Prices")

# Save the plot
plt.savefig("kdeplot_price_distribution.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The KDE plot shows a smooth density curve representing the distribution of diamond prices.

Key insights:
- The highest density occurs at lower price values.
- The curve gradually tapers off toward higher prices, confirming right skewness.
- KDE provides a clearer understanding of distribution shape compared to histograms alone.

This visualization effectively highlights the overall price distribution of diamonds.


### üîπ jointplot(): Investigating Bivariate Relationships

The `jointplot()` function in Seaborn is used to visualize the relationship between two numerical variables along with their individual distributions.  
It combines a central scatter plot with marginal plots (histograms or KDE plots) on the axes.

In this visualization, we analyze the relationship between:
- **carat**: Weight of the diamond  
- **price**: Price of the diamond  

This helps in understanding both the correlation and the distribution of these variables simultaneously.


```python
g = sns.jointplot(
    x="carat",
    y="price",
    data=df,
    kind="hex",
    cmap="viridis"
)

g.fig.suptitle("Joint Plot of Carat and Price with Density Gradient", y=1.02)

# Save the plot
g.savefig("jointplot_carat_price_gradient.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The hexbin joint plot uses a color gradient to represent the density of data points.  
Darker regions indicate areas with a higher concentration of diamonds.

Key insights:
- Most diamonds are concentrated at lower carat values and lower prices.
- As carat increases, price increases sharply, but data density decreases.
- The gradient helps clearly identify regions with high and low data concentration.

This visualization effectively combines relationship analysis with density-based insight.


### üîπ rugplot(): Enhancing Visualizations with Rug Plots

The `rugplot()` function in Seaborn is used to display individual data points along an axis.  
It is often used alongside other distribution plots to show the exact location of observations.

In this visualization, we use a rug plot to enhance the understanding of the **distribution of diamond prices** by showing individual price values along the axis.


```python
plt.figure(figsize=(8, 2))

sns.rugplot(
    df["price"],
    height=0.5,
    color="purple"
)

plt.title("Rug Plot of Diamond Prices")

# Save the plot
plt.savefig("rugplot_price_distribution.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The rug plot displays individual diamond prices as small ticks along the axis.

Key insights:
- A dense concentration of ticks appears at lower price values.
- Fewer ticks are present at higher prices, indicating fewer high-priced diamonds.
- The rug plot complements histogram and KDE plots by showing exact data points.

This visualization provides a precise view of data distribution at the individual observation level.

## üìâ Linear Regression and Relationships

Linear regression helps in understanding the relationship between two numerical variables by fitting a best-fit line through the data points.  
Seaborn provides simple yet powerful functions to perform and visualize linear regression analysis.

In this section, we explore linear relationships using:
- **regplot()**
- **lmplot()**

We begin with `regplot()`, which is used to visualize a regression line along with scatter points.


### üîπ regplot(): Performing Linear Regression Analysis

The `regplot()` function in Seaborn is used to plot data points along with a fitted linear regression line.  
It helps in understanding the strength and direction of the relationship between two numerical variables.

In this visualization, we perform linear regression between:
- **carat**: Weight of the diamond  
- **price**: Price of the diamond  

This analysis helps quantify how diamond price changes with respect to carat weight.


```python
plt.figure(figsize=(8, 5))

sns.regplot(
    x="carat",
    y="price",
    data=df,
    scatter_kws={"alpha": 0.4},
    line_kws={"color": "red"}
)

plt.title("Linear Regression Between Carat and Price")

# Save the plot
plt.savefig("regplot_carat_price.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The regression plot shows a strong positive linear relationship between **carat** and **price**.

Key observations:
- The upward-sloping regression line indicates that price increases as carat increases.
- The scatter points show some variability, especially at higher carat values.
- Despite the non-linear nature of price growth, the linear trend captures the overall direction of the relationship.

This plot confirms that **carat is a significant predictor of diamond price**.

### üîπ lmplot(): Visualizing Linear Relationships Between Variables

The `lmplot()` function in Seaborn is a high-level interface used to visualize linear relationships between variables.  
Unlike `regplot()`, `lmplot()` supports grouping using the `hue` parameter and can create multiple regression plots based on categorical variables.

In this visualization, we analyze the linear relationship between:
- **carat**: Weight of the diamond  
- **price**: Price of the diamond  

We also group the data based on **cut** to observe how the relationship varies across different cut categories.

```python
g = sns.lmplot(
    x="carat",
    y="price",
    data=df,
    hue="cut",
    palette="viridis",
    scatter_kws={"alpha": 0.3},
    height=5,
    aspect=1.4
)

g.fig.suptitle("Linear Relationship Between Carat and Price by Cut", y=1.02)

# Save the plot
g.savefig("lmplot_carat_price_by_cut.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The lmplot shows linear regression lines for different diamond cut categories.

Key insights:
- All cut categories show a positive relationship between carat and price.
- Higher quality cuts tend to have steeper regression lines, indicating higher prices for the same carat.
- Variations between regression lines highlight how cut quality influences pricing.

This visualization demonstrates how categorical variables can affect linear relationships in data.

## üîó Correlation Heatmap & Pairwise Relationships

Correlation analysis helps in understanding the strength and direction of relationships between numerical variables.  
Pearson correlation is commonly used to measure linear relationships between continuous features.

In this section, we:
- Compute Pearson correlation coefficients between numerical features
- Visualize correlations using a heatmap
- Use masking and annotation for better readability
- Analyze pairwise relationships
- Identify the strongest positive and negative correlations


### üîπ Selecting Numerical Features

Correlation analysis is applicable only to numerical data.  
Therefore, we first extract all numerical columns from the Diamonds dataset to compute correlations.


```python
numeric_df = df.select_dtypes(include=["int64", "float64"])
numeric_df.head()
```

### üîπ Computing Pearson Correlation Matrix

The Pearson correlation coefficient measures the linear relationship between two numerical variables.  
Its value ranges from:
- **+1** (strong positive correlation)
- **0** (no correlation)
- **‚àí1** (strong negative correlation)

We compute the Pearson correlation matrix for all numerical features.


```python
corr_matrix = numeric_df.corr(method="pearson")
corr_matrix
```

### üîπ Correlation Heatmap Visualization

A heatmap provides a visual representation of correlation values using color gradients.  
To avoid redundancy, we mask the upper triangle of the correlation matrix and annotate each cell with its correlation value.


```python
import numpy as np

# Create mask for upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap of Numerical Features")

# Save the heatmap
plt.savefig("correlation_heatmap.png", dpi=300, bbox_inches="tight")

plt.show()
```

### üìå Interpretation

The heatmap reveals several strong correlations among numerical features:

- **carat and price** show a strong positive correlation, indicating that price increases as carat weight increases.
- **x, y, and z** dimensions are highly correlated with each other, as they represent physical measurements of diamonds.
- **depth and table** show weaker correlations with price, suggesting a lesser influence on pricing.

The heatmap effectively highlights the strength and direction of relationships between numerical variables.


### üîπ Pairwise Relationships Using Pairplot

Pairplots (scatter matrix) allow simultaneous visualization of pairwise relationships between multiple numerical variables.  
They help in identifying trends, correlations, and potential outliers across features.


```python
sns.pairplot(
    df.sample(500),          # use full df to allow hue
    hue="cut",               # categorical color grouping
    palette="viridis",
    diag_kind="kde"
)

plt.savefig("pairplot_numerical_features_colored.png", dpi=300, bbox_inches="tight")
plt.show()
```

The colored pairplot highlights pairwise relationships between numerical variables while grouping data by diamond cut.  
Different color clusters show how numerical relationships vary across cut categories, enhancing interpretability.


Note: Pairplots require a categorical variable to apply color palettes.
Numeric-only datasets do not support color gradients directly in pairplots.

### üîπ Strongest Positive and Negative Relationships

To identify the strongest relationships, we examine the correlation values numerically.


```python
corr_pairs = corr_matrix.unstack().sort_values(ascending=False)

# Remove self-correlation
corr_pairs = corr_pairs[corr_pairs < 1]

corr_pairs.head(10)
```

### üìå Correlation Summary

- The strongest positive correlation is observed between **carat and price**, indicating that carat weight is the most influential factor in diamond pricing.
- Strong positive correlations also exist between **x, y, and z**, reflecting the physical dimensions of diamonds.
- No strong negative correlations are observed among the numerical features.
- Some variables show weak correlations, indicating limited linear dependency.

Overall, the correlation analysis provides valuable insights into how numerical attributes interact within the dataset.
